### 3. Gradient boosting machine
##### working on it

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
# first use the basic no text data
data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset_no_text.csv")
test_data = pd.read_csv("../COMP30027_2024_asst2_data/test_dataset_no_text.csv")
# drop id before calculation, add back after
data = data.drop(columns='id')
test_data = test_data.drop(columns='id')
```

```python
# split the data
X = data.drop(columns=["imdb_score_binned"])
y = data["imdb_score_binned"]

# set up scalers
mmscaler = MinMaxScaler()
ssscaler = StandardScaler()

mmscaler.fit(X)
ssscaler.fit(X)

mmX = mmscaler.transform(X)
ssX = ssscaler.transform(X)

gbm = GradientBoostingClassifier()

scores_minmax = cross_val_score(gbm, mmX, y, cv=10)
scores_standard = cross_val_score(gbm, ssX, y, cv=10)

print("Scores with Min-Max scaling:", scores_minmax.mean())
print("Scores with Standardization:", scores_standard.mean())
```

#### Kaggle prediction on test data

```python
# make prediction on test
# despite the min-max scaling perform slightly better
# the large scale of dataset with outlier should perform best
# with standardization

test_data = ssscaler.fit_transform(test_data)

gbm.fit(ssX, y)

test_pred = gbm.predict(test_data)
ids = range(1, len(test_pred) + 1)

# save submission file
prediction = pd.DataFrame({'id': ids, 'imdb_score_binned': test_pred})
prediction.to_csv("../Kaggle_submissions/GBM_no_text.csv", index=False)
```
