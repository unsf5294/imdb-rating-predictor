### 2. Support Vector Machine
##### This model is a different approach to K-NN model, as both involve distance calculation in dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
```

#### 2.1 SVM using a simple preprocess with low dimensionality and many columns removed.
##### AKA data_no_text

```python
# first use the basic no text data
data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset_elim.csv")
test_data = pd.read_csv("../COMP30027_2024_asst2_data/test_dataset_elim.csv")
# drop id before calculation, add back after
data = data.drop(columns='id')
test_data = test_data.drop(columns='id')
```

```python
# split the data
X = data.drop(columns=["imdb_score_binned"])
y = data["imdb_score_binned"]

# set up scalers
mmscaler = MinMaxScaler()
ssscaler = StandardScaler()

mmscaler.fit(X)
ssscaler.fit(X)

mmX = mmscaler.transform(X)
ssX = ssscaler.transform(X)


svm = SVC()

scores_minmax = cross_val_score(svm, mmX, y, cv=10)
scores_standard = cross_val_score(svm, ssX, y, cv=10)

print("Scores with Min-Max scaling:", scores_minmax.mean())
print("Scores with Standardization:", scores_standard.mean())
```

#### Kaggle prediction on test data

```python
test_data = ssscaler.fit_transform(test_data)

svm.fit(ssX, y)

test_pred = svm.predict(test_data)
ids = range(1, len(test_pred) + 1)

# save submission file
prediction = pd.DataFrame({'id': ids, 'imdb_score_binned': test_pred})
prediction.to_csv("../Kaggle_submissions/SVM_no_text.csv", index=False)
```

#### 2.2 SVM using dataset with genre and plot keywords embedding

```python
data = pd.read_csv("../COMP30027_2024_asst2_data/train_pg_PCA.csv")
data = data.drop(columns='id')
```

```python
# split the data
X = data.drop(columns=["imdb_score_binned"])
y = data["imdb_score_binned"]

# set up scalers
mmscaler = MinMaxScaler()
ssscaler = StandardScaler()

mmscaler.fit(X)
ssscaler.fit(X)

mmX = mmscaler.transform(X)
ssX = ssscaler.transform(X)

svm = SVC()

scores_minmax = cross_val_score(svm, mmX, y, cv=10)
scores_standard = cross_val_score(svm, ssX, y, cv=10)

print("Scores with Min-Max scaling:", scores_minmax.mean())
print("Scores with Standardization:", scores_standard.mean())
# due to the numeric nature of SVM, a dataset with high dimensionality might not work well
# this is due to the massive columns that related to plot and genre 'overweight' the rest of columns by
# take the majority of feature space.
```
